```python
from scipy.linalg import lu as LU
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
def create_A_b(shape=(2, 2), k=1):
    A = np.random.rand(*shape) * k
    b = np.random.rand(shape[0]) * k
    return A, b
```

We should solve the system equation:  
Ly = b  
Ux = y  
or:  
L_inv * L * y = L_inv * b -> y = L_inv * b  
U_inv * U * x = U_inv * y -> x = U_inv * y

```python
def find_x(A, b):
    # LU разложение
    def create_LU(A):
        L, U = LU(A, permute_l=True)
        assert np.allclose(L @ U, A), ValueError("Not correct LU")
        return L, U

    L, U = create_LU(A)
    # обратные матрицы
    L_inv = np.linalg.inv(L)
    U_inv = np.linalg.inv(U)
    # новый вектор y
    y = L_inv @ b
    # искомый вектор x
    x = U_inv @ y
    #print('Comparing results:')
    #print("A * x =\t", A @ x)
    #print("b =\t", b)
    mse = ((A @ x - b)**2).mean(axis=-1)
    mae = np.abs((A @ x - b).mean(axis=-1))
    #print("MSE:\t", mse)
    #print("MAE:\t", mae)
    return x, mse, mae
```

```python
x, mse, mae = find_x(*create_A_b((100, 100), 1))
```

```python
MSE = {}
MAE = {}
for k in tqdm(range(1000, 101001, 5000)):
    MSE[k] = []
    MAE[k] = []
    for size in range(2, 11):
        x, mse, mae = find_x(*create_A_b((size*10, size*10), k))
        MSE[k] += [mse]
        MAE[k] += [mae]
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('LU')
plt.plot(MSE[101000])
plt.xlabel('Size x10')
plt.ylabel('MSE')
plt.subplot(1, 2, 2)
plt.title('LU')
plt.plot(MAE[101000])
plt.xlabel('Size x10')
plt.ylabel('MAE')
plt.show()
```

Метод вариаций

```python
def solve_with_rotation(A, b):
    m = np.concatenate((A, np.expand_dims(b, 1)), axis=-1)
    n = m.shape[0]
    # forward trace
    for i in range(n-1):
        for j in range(i + 1, n):
            c = m[i, i] / (m[i, i]**2 + m[j, i]**2) ** .5
            s = m[j, i] / (m[i, i]**2 + m[j, i]**2) ** .5
            tmp1 = m[i, :] * c + m[j, :] * s
            tmp2 = m[i, :] * -s + m[j, :] * c
            m[i, :] = tmp1
            m[j, :] = tmp2
    
    # find x, backward trace
    x = np.array([0.0 for i in range(n)])
    for k in range(n - 1, -1, -1):
        x[k] = (m[k, -1] - m[k, k:n] @ x[k:n]) / m[k, k]

    mse = ((A @ x - b)**2).mean(axis=-1)
    mae = np.abs((A @ x - b).mean(axis=-1))
    return x, mse, mae
```

```python
MSE = {}
MAE = {}
for k in tqdm(range(1000, 101001, 5000)):
    MSE[k] = []
    MAE[k] = []
    for size in range(2, 11):
        x, mse, mae = solve_with_rotation(*create_A_b((size*10, size*10), k))
        MSE[k] += [mse]
        MAE[k] += [mae]
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 2, 1)
plt.title('Rotation')
plt.plot(MSE[101000])
plt.xlabel('Size x10')
plt.ylabel('MSE')
plt.subplot(1, 2, 2)
plt.title('Rotation')
plt.plot(MAE[101000])
plt.xlabel('Size x10')
plt.ylabel('MAE')
plt.show()
```
