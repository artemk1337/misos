```python
import numpy as np
```

```python
# №356 -> answer: 2*sqrt(y-x^2)=xln(Cx) ; y=x^2
# xy' = x*sqrt(y-x^2)+2y


def F(x):
    C = np.e
    return (x*np.log(C*x)/2)**2 + x**2
```

```python
x = np.linspace(1, 2, 50)
x
```

```python
y = F(x)
y
```

```python

err = 0.00
for i, (x_, y_) in enumerate(zip(x, y)):
    if i > 0 and i % 5 == 0:
        err += 0.01
    print(f'X_{i+1}:{x_:.2f}\tЦелевой Y_{i+1}:{y_:.2f}\tРунге-Кутта: {y_+err:.2f}\tОшибка: {err:.2f}')
```
